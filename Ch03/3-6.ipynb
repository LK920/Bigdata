```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#기본 플롯 차트
plt.title('plot1')
plt.plot([1,3,4,9,10,7])
plt.show()
```

```python
#x축 지정
plt.title('plot2')
plt.plot([10, 20,30,40,50,60],[1,3,4,9,10,7])
plt.show()
```

```python
#차트 스타일
plt.title('plot3')
plt.plot([10, 20,30,40,50,60],[1,3,4,9,10,7],'go:')
plt.grid(True)
plt.show()
```

```python
#멀티 라인차트 -plot를 계속 추가하면 된다.
data = np.arange(0, 5, 0.2)
data

plt.title('plot4')
plt.plot(data, data, 'r--')
plt.plot(data, data**2, 'b--')
plt.plot(data, data**3, 'g--')
plt.show()
```

```python
#막대 그래프
plt.title('plot5')
plt.bar(['apple','banana','orange','grape','barry'],
        [5,1,7,5, 15])
plt.xlabel('fruit')
plt.ylabel('count')
plt.show()
```

```python
#히스토그램(연속값을 출력할 때 사용되는 막대그래프)
data1 = np.random.rand(1000)
data2 = np.random.randn(1000)

plt.title('hist1')
plt.hist(data1, edgecolor='black')
plt.show()
```

```python
plt.title('hist1')
plt.hist(data2, edgecolor='black')
plt.show()
```

```python
#산점도
x = np.random.normal(0,1, 1000)
y = np.random.normal(0,1, 1000)

plt.title('scatter')
plt.scatter(x,y)
plt.show()
```
