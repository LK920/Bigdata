```python
"""
date : 20/08/12
name : kang
content : 머신러닝 - 군집화 실습하기
"""
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
#데이터 프레임 생성
df = pd.read_csv('./data/iris.csv')
df_iris = df.iloc[:,0:4]
df_iris
```

```python
#k-means 알고리즘으로 군집화
kmeans = KMeans(n_clusters=3).fit(df_iris.values)
```

```python
#군집화 결과확인
df_iris['cluster_id'] = kmeans.labels_
df_iris
```

```python
#그룹화 결과 시각화
sns.lmplot('sepal.length',
           'sepal.width',
           data=df_iris,
           fit_reg=False,
           scatter_kws={'s':150},
           hue='cluster_id'
          )
```
